# Pregunta 2
***Los data frame df1 y df2 son parte de un data frame llamado df_solicitudes, el postulante debe generar ese data frame a traves de la uni√≥n de estos dos data frame donde la columna para efectuar la relaci√≥n entre ambos es "CODSOLICITUD" y que solo debe tomarse las columnas "CODMES", "DEUDA_TOTAL_SISTEMA", "CLASIF_SISTEMA_ULT_12M" del data frame df2***

```python
# Reordenar las columnas de df2 para que "CODSOLICITUD" est√© en la misma posici√≥n que en df1
columnas_df2 = ['CODMES','CODSOLICITUD', 'DEUDA_TOTAL_SISTEMA', 'CLASIF_SISTEMA_ULT_12M']
# Unir df1 y df2 en base a la columna "CODSOLICITUD" y seleccionar solo las columnas deseadas de df2
df_solicitudes_new = pd.merge(df1, df2[columnas_df2], on='CODSOLICITUD', how='inner')
df_solicitudes_new.head()
```
